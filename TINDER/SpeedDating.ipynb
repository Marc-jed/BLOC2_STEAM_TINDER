## Les gens peuvent-ils prdire avec prcision leur propre valeur perue sur le march des rencontres ?
#### Nous allons crer une nouvelle colonne afin de calculer un indice de corrlation permettant de voir si les candidats ont une valuation proche de l'valuation du Date rencontr. Pour cela nous utiliserons les colonnes attr_o ... amb_o correspondant aux notes donnes par le Date et attr5_1 ... amb5_1 correspondant aux notes donnes soit mme.


```python
# on garde que les colonnes necessaires  notre calcule
cols = ['iid', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 
        'attr5_1', 'sinc5_1', 'intel5_1', 'fun5_1', 'amb5_1']
# on supprime les NaN
df_corr = df[cols].dropna()

# On cre une fonction pour calculer la corrlation entre les attributs
def compute_corr(row):
    date = [row['attr_o'], row['sinc_o'], row['intel_o'], row['fun_o'], row['amb_o']]
    perso = [row['attr5_1'], row['sinc5_1'], row['intel5_1'], row['fun5_1'], row['amb5_1']]
    # l'utilisation dela fonction corr necessite d'utiliser des donnes en Series
    return pd.Series(date).corr(pd.Series(perso))

# Appliquer  chaque ligne
df_corr['alignement_corr'] = df_corr.apply(compute_corr, axis=1)

# Rsum statistique
print(df_corr['alignement_corr'].describe(include='all'))
```
