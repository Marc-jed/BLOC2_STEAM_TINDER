# Commençons à explorer la donnée

#### Nous allons créer un dataset n'ayant que les iid unique afin d'avoir bien des données de participant unique
#### Nous prendrons le temps au fur et à mesure de modifier certaines colonne pour une meilleur lecture des variables

```python
# Modificationde colonne pour une meilleur compréhension des valeurs
df["dec_o"] = df["dec_o"].apply( lambda x : "Non" if x==0 else "Oui" if x==1 else x)
df["gender"] = df["gender"].apply( lambda x : "Woman" if x==0 else "Man" if x==1 else x)
df["samerace"] = df["samerace"].apply(lambda x : "Non" if x==0 else "Oui" if x==1 else x)
df["dec"]=df["dec"].apply(lambda x: "Non" if x==0 else "Oui" if x==1 else x)
df["date_3"]=df["date_3"].apply(lambda x:"Non" if x==0 else "Oui" if x==1 else x)

df.head()
```

```python
df_unique = df.drop_duplicates(subset='iid')
```

```python
df_unique.head()
```

```python
# configuration d'un code couleur pour Woman et Man
color_discrete_map = {
    'Man': 'blue',  
    'Woman': 'red'  
}
# Répartition Femme et Homme
id_unique = df.groupby('gender')['iid'].nunique().reset_index()

fig=px.bar(id_unique, x="gender", y="iid", color="gender", text="iid", color_discrete_map=color_discrete_map, title="Nombre de Femme et d'Homme participant à cette étude")
fig.show()


fig2 = px.histogram(df_unique, x="age",color="gender",
                    marginal="box",
                    color_discrete_map=color_discrete_map,
                    title="Répartition de l'age des candidats"
                    )
fig2.show()
```

```python
# Quelques données liés à l'age de nos candidats
mask_age = (df_unique["age"] >= 23) & (df_unique["age"] <= 29)
df_unique_age = df_unique[mask_age]
distribution = df_unique_age["age"].count()
pourcentage_distribution = distribution*100/df_unique["age"].count()
print("L'age des candidats est compris entre 23 et 29 ans:", distribution )
print(f"Soit {pourcentage_distribution.round(2)}%")
mask_gender = df_unique["gender"] == "Woman"
df_unique_wgender = df_unique[mask_gender]
woman_mean = df_unique_wgender["age"].mean()
df_unique_mgender= df_unique[~mask_gender]
man_mean = df_unique_mgender["age"].mean()
print(f"L'age moyen des femmes est de {woman_mean.round(1)} ans, et l'age moyen des hommes est de {man_mean.round(1)} ans")
```

#### La distribution de l'age montre que 2/3 des candidats sont agés entre 23 et 29 ans et qu'il y a quelques outliers pour les 2 partis. 
#### Les hommes présentent un age avec une distribution plus grande que les femmes et donc un age moyen plus élevé. Quand aux femmes elles couvrent principalement une plage d'age plus petit mais a la candidate la plus agée.

## Les gens peuvent-ils prédire avec précision leur propre valeur perçue sur le marché des rencontres ?
#### Nous allons créer une nouvelle colonne afin de calculer un indice de corrélation permettant de voir si les candidats ont une évaluation proche de l'évaluation du Date rencontré. Pour cela nous utiliserons les colonnes attr_o ... amb_o correspondant aux notes données par le Date et attr5_1 ... amb5_1 correspondant aux notes données soit même.


```python
# on garde que les colonnes necessaires à notre calcule
cols = ['iid', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 
        'attr5_1', 'sinc5_1', 'intel5_1', 'fun5_1', 'amb5_1']
# on supprime les NaN
df_corr = df[cols].dropna()

# On crée une fonction pour calculer la corrélation entre les attributs
def compute_corr(row):
    date = [row['attr_o'], row['sinc_o'], row['intel_o'], row['fun_o'], row['amb_o']]
    perso = [row['attr5_1'], row['sinc5_1'], row['intel5_1'], row['fun5_1'], row['amb5_1']]
    # l'utilisation dela fonction corr necessite d'utiliser des données en Series
    return pd.Series(date).corr(pd.Series(perso))

# Appliquer à chaque ligne
df_corr['alignement_corr'] = df_corr.apply(compute_corr, axis=1)

# Résumé statistique
print(df_corr['alignement_corr'].describe(include='all'))
```

```python
# Sélectionner uniquement les colonnes utiles
cols = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 
        'attr5_1', 'sinc5_1', 'intel5_1', 'fun5_1', 'amb5_1']

# Supprimer les NaN
df_corr = df[cols].dropna()


corr_matrix = df_corr.corr()

fig = px.imshow(
    corr_matrix,
    text_auto=True,         
    color_continuous_scale="RdBu_r", 
    zmin=-1, zmax=1,         
    title="Matrice de corrélation entre auto-évaluation et notes données"
)

fig.update_layout(
    width=800, height=600,
    xaxis_title="Variables",
    yaxis_title="Variables"
)

fig.show()
```

Les corrélations entre auto-évaluation et notes données sont proches de 0, donc très faibles. Cela signifie que les participants n’arrivent pas à bien se juger, ou que l’image qu’ils ont d’eux-mêmes ne correspond pas à la réalité perçue par les autres. On observe en revanche des corrélations entre certaines notes : lorsqu’une personne s’auto-évalue comme sincère, elle a tendance à se juger aussi plus intelligente. Côté évaluation des partenaires, on retrouve le même phénomène : une personne jugée attirante est souvent perçue aussi comme drôle ou intelligente

Concernant la dernière note tous les critères chutent sauf le "drole/amusant" qui remonte en flèche. Le dernier Date a du être moins pertinent dans la décision du match car certainement que seul ce dernier critère a du faire pencher la balance.

# Pour conclure
Le dataset contient de nombreuses valeurs manquantes, ce qui a nécessité un traitement attentif.

Les critères d’évaluation diffèrent au début entre hommes et femmes, mais finissent par converger vers deux attributs centraux :

l’attirance,

l’intelligence.

L’attirance, peu prioritaire pour les femmes au départ (contrairement aux hommes), prend progressivement de l’importance au fil de la soirée.

Pour les hommes, c’est l’intelligence qui gagne en poids décisionnel.

L’appartenance à la même race n’est pas un facteur déterminant. En revanche, le partage d’intérêts communs améliore légèrement les chances de match.

Les candidats peinent globalement à s’auto-évaluer sur les attributs, même si certains y parviennent mieux que d’autres.

Être parmi les premiers rendez-vous offre une très légère probabilité supplémentaire d’obtenir un match. En fin de soirée, les taux de match deviennent plus instables, avec de fortes variations positives et négatives, possiblement liées à la fatigue ou à une baisse d’attention.
